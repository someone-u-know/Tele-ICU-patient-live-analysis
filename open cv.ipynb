```python

import cv2
import numpy as np
import os
import schedule
import time

# Define the video path
video_link = 'vid3.mp4'  # Replace with your video path

# Create a folder to store the frames
frame_folder = 'frames'
if not os.path.exists(frame_folder):
    os.makedirs(frame_folder)

# Create a VideoCapture object
cap = cv2.VideoCapture(video_link)

if not cap.isOpened():
    print("Error: Could not open video file.")
    exit()

def extract_frame():
    global cap  # Use the global cap object
    ret, frame = cap.read()
    if ret:
        frame = preprocess_frame(frame)
        # Store the preprocessed frame in the folder
        frame_path = os.path.join(frame_folder, f'frame_{int(time.time())}.jpg')
        cv2.imwrite(frame_path, frame)
        print("Frame extracted, preprocessed, and saved!")
    else:
        print("No more frames to read or error reading frame from video file.")
        schedule.clear()  # Stop the scheduling
        cap.release()  # Release the VideoCapture object

def preprocess_frame(frame):
    frame = cv2.resize(frame, (224, 224))
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    return frame

schedule.every(1).seconds.do(extract_frame)  # Extract a frame every 10 seconds

while True:
    schedule.run_pending()
    time.sleep(1)
```
